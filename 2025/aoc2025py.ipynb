# Day 1

Something about Locked doors and codes... or somethig like that  
https://adventofcode.com/2025/day/1


```python
def day1():
    file = open('input.txt', 'r').readlines()
    start = 50
    zeros_count = 0
    for line in file:
        amount = line[1:].strip()
        if line[0]=='L':
            start = (start + 100 - int(amount))% 100
        elif line[0]=='R':
            start = (start + int(amount))%100
        if start==0:
            zeros_count += 1
    print("Parte 1 (número de zeros):", zeros_count)

    start = 50
    zeros_count = 0
    for line in file:
        amount = int(line[1:].strip())
        if line[0]=='L':
            while amount>0:
                start = (start + 100 - 1) % 100
                amount-=1
                if start==0:
                    zeros_count += 1
        elif line[0]=='R':
            while amount>0:
                start = (start + 1 ) % 100
                amount-=1
                if start==0:
                    zeros_count += 1
    print("Parte 2 (número de clickes no zero):", zeros_count)

day1()
```

Something about a gift shop smells RE  
https://adventofcode.com/2025/day/2

```python
def is_valid_id(id, retxt):
    """What is a valid id? something that doesn't fit a """
    import re
    p = re.compile(retxt)
    if p.findall(str(id)):
        return id 
    else: 
        return 0
```

```python
data = "11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"
elms = [el.split("-") for el in data.split(",")]
```

```python
total1 = 0
total2 = 0
for el in elms:
    a = int(el[0])
    b = int(el[1])
    for t in range(a,b+1):
        total1 += is_valid_id(t, r"^(\d+)(\1)$")
        total2 += is_valid_id(t, r"^(\d+)(\1)+$")

print("Parte 1:", total1)
print("Parte 2:", total2)
```

# Day 3

--- Day 3: Lobby ---  
https://adventofcode.com/2025/day/3

```python
data = """987654321111111
811111111111119
234234234234278
818181911112111""".split("\n")
```

```python
def rec_jolt(bank, current, n_bateries):
    """Recursive solution for any number of bateries"""
    if n_bateries==0:
        return current
    ni = 0
    for i in range(len(bank)-n_bateries+1):
         if int(bank[i])>int(bank[ni]):
             ni = i
    return rec_jolt(bank[(ni+1):], current+bank[ni], n_bateries-1)
```

```python
total1 = 0
total2 = 0
for bank in data:
    total1 += int(rec_jolt(bank, "", 2))
    total2 += int(rec_jolt(bank, "", 12))
print("Parte 1:", total1)
print("Parte 2:", total2)
```

--- Day 4: Printing Department ---  
https://adventofcode.com/2025/day/4

```python
data="""..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.""".split("\n")
```

```python
total1 = 0
total2 = 0
cols = len(data[0])
rows = len(data)

def count_vizinhos(r,c, data):
    dd = [-1, 0, 1]
    cols = len(data[0])
    rows = len(data)
    vizinhos = -1
    for dy in dd:
        for dx in dd:
            if 0 <= r + dy < rows and 0 <= c + dx < cols:
                if data[r+dy][c+dx] == "@":
                    vizinhos += 1
    return vizinhos

for it in range(10_000): # make sure we can get out of a loop.
    mark = []
    for r in range(rows):
        for c in range(cols):
            if data[r][c] == '@':
                vizinhos = count_vizinhos(r,c,data)
                if vizinhos < 4:
                    mark.append((r,c))
                    total1 += 1
    if (it == 0):
        print("Parte 1:", total1)
    for m in mark:
        r = m[0]
        c = m[1]
        data[r]= data[r][:c] + "." + data[r][(c+1):]
    if not mark:
        break
print("Parte 2:", total1)
```

```python
# count ids, better way. Merge intervals, count totals by S=top-bottom+1

def merge_two(a,b):
    """Merge two ranges, used by merge_list"""
    if b[1] < a[0] or b[0] > a[1]:
        return [a,b]
    return [[min(a[0], b[0]), max(a[1], b[1])]]

def merge_list(a):
    # Identify the ranges to remove by adding new ones
    to_remove = set()
    to_add = []
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            t = merge_two(a[i],a[j])
            # print("merged", t, len(t))
            if len(t)==1:
                to_add.extend(t)
                to_remove.add(i)
                to_remove.add(j)
            if len(t)==2:
                pass

    # Remove ranges that were merged
    merged=a[:]
    for i in to_remove:
        while a[i] in merged:
            merged.remove(a[i])
    
    # Add the ranges that were created
    merged.extend(to_add)

    # Remove any duplicates in the list
    for i in range(len(merged)-1, 0, -1):
        if merged[i] in merged[:i]:
            merged.remove(merged[i])

    return merged if merged else a
```

```python
# Repeat the merging of the list until there's no change in the output
for it in range(500):
    b = merge_list(valid)
    if b==valid:
        print("No repeats at it", it)
        break
    valid = b

# Sum each range sum it (top-bottom-1)
total2 = 0
for el in valid:
    total2 += el[1]-el[0]+1

print("Parte 2:", total2)
```

```python
data2 = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """.split("\n")
data = [ a.split() for a in data2]
```

```python
acum = [1 if a == '*' else 0 for a in data[-1]]
for i in range(len(data)-1):
    for j in range(len(data[i])):
        if data[-1][j] == '*':
            acum[j] *= int(data[i][j])
        if data[-1][j] == '+':
            acum[j] += int(data[i][j])
print("Parte 1: ", sum(acum))
```

```python
data2 = open("day6.txt", "r").read()
data2 = data2.split("\n")
```

```python
oper = data2[-1].split()
nums = data2[:-1]
problem = len(oper)-1
acum = [1 if a == '*' else 0 for a in oper]
for i in range(len(nums[0])-1,-1,-1):
    if nums[0][i] == ' ' and nums[1][i] == ' ' and nums[2][i] == ' ' and nums[3][i]== ' ':
        problem -= 1
        continue
    col = "".join([nums[0][i], nums[1][i], nums[2][i], nums[3][i]])      
    if oper[problem] == "*":
        acum[problem] *= int(col)
    if oper[problem] == "+":
        acum[problem] += int(col)
    
print("Parte 2:", sum(acum))
```
