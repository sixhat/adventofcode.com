https://adventofcode.com/2025/day/1


```python
def day1():
    file = open('input.txt', 'r').readlines()
    start = 50
    zeros_count = 0
    for line in file:
        amount = line[1:].strip()
        if line[0]=='L':
            start = (start + 100 - int(amount))% 100
        elif line[0]=='R':
            start = (start + int(amount))%100
        if start==0:
            zeros_count += 1
    print("Parte 1 (número de zeros):", zeros_count)

    start = 50
    zeros_count = 0
    for line in file:
        amount = int(line[1:].strip())
        if line[0]=='L':
            while amount>0:
                start = (start + 100 - 1) % 100
                amount-=1
                if start==0:
                    zeros_count += 1
        elif line[0]=='R':
            while amount>0:
                start = (start + 1 ) % 100
                amount-=1
                if start==0:
                    zeros_count += 1
    print("Parte 2 (número de clickes no zero):", zeros_count)

day1()
```
